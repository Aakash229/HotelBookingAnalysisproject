BB: Bed and Breakfast

This typically includes accommodations (bed) and breakfast. Guests receive a room for the night and breakfast the following morning. Other meals are not included.

HB: Half Board

This typically includes accommodations, breakfast, and one other meal, usually dinner. Guests receive a room for the night, breakfast, and dinner, but lunch is not included.

SC: Self-Catering

This means guests have access to cooking facilities within their accommodation and are responsible for preparing their own meals. No meals are provided by the hotel.
FB: Full Board

This typically includes accommodations, breakfast, lunch, and dinner. Guests receive a room for the night, as well as all three main meals of the day.
